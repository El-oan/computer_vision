ImageEncoder (Hiera): A hierarchical Masked Autoencoder (MAE) pre-trained vision transformer. Unlike the original SAM's single-scale backbone, this extracts multi-scale features.

MemoryEncoder & MemoryBank: A memory mechanism that stores information from past frames. It maintains a First-In-First-Out (FIFO) queue of up to N=6 recent unprompted frames and M=1 prompted frames.

MemoryAttention: A module containing L=4 transformer blocks that uses cross-attention with Rotary Position Embedding (RoPE) to condition the current frame's features on the past frames stored in the memory bank.

PromptEncoder: A module that takes user inputs (clicks, boxes, masks, or text) and projects them into prompt tokens.
MaskDecoder: A decoder that integrates the image features, prompt tokens, and memory-conditioned features via cross-attention to output the final segmentation masks and an explicit occlusion score
